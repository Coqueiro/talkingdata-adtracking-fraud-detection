# Fraud risk #

Fraud risk is everywhere, but for companies that advertise online, click fraud can happen at an overwhelming volume, resulting in misleading click data and wasted money. Ad channels can drive up costs by simply clicking on the ad at a large scale. With over 1 billion smart mobile devices in active use every month, China is the largest mobile market in the world and therefore suffers from huge volumes of fradulent traffic.

TalkingData, China’s largest independent big data service platform, covers over 70% of active mobile devices nationwide. They handle 3 billion clicks per day, of which 90% are potentially fraudulent. Their current approach to prevent click fraud for app developers is to measure the journey of a user’s click across their portfolio, and flag IP addresses who produce lots of clicks, but never end up installing apps. With this information, they've built an IP blacklist and device blacklist.

While successful, they want to always be one step ahead of fraudsters and have turned to the Kaggle community for help in further developing their solution. In their 2nd competition with Kaggle, you’re challenged to build an algorithm that predicts whether a user will download an app after clicking a mobile app ad. To support your modeling, they have provided a generous dataset covering approximately 200 million clicks over 4 days!

For this competition, your objective is to predict whether a user will download an app after clicking a mobile app advertisement.

# Considerations #

Although the main theme of the challenge is fraudulent behaviour, we don't need to deal with this concept, since we are trying to predict for downloads after clicks or not. Of course, it's expected that a trained model will give more discredit to a fraudulent behaviour, but that won't be necessarily the case.

We are dealing with millions of rows, and thus processing can be very heavy on computational resources. For EDA we are not going to use the full training set, but rather a subset. While the full training subset may help prediction scores, a smaller subset may be easier to maneuver, and with enough size, shouldn't debilitate the prior analysis.

This is a classification problem with two possible classes, and we will soon observe that we have a very skewed attribution pattern: most of the clicks don't result in an app download (`is_attributed = 1`). EDA has an important function of revealing observable patterns, calculating baselines and improve intuition on the data.

Another detail: although encoded columns have `int` values, we should treat them as categorical variables, which is closer to their real representation.

We may load a random subset from the `train.csv` using `dask`. `train.csv` is ordered by click date, so we can't just collect first N rows.

````python
start_time = time.time()

# Loading data #
df = dd.read_csv('data/train.csv', parse_dates=['click_time', 'attributed_time'])
freq = 0.01 # 1% of train dataset
df = df.random_split([freq , 1-freq])[0]
df = df.compute()

print('[{}] Finished to load data'.format(time.time() - start_time))
df.info()
````

## Data fields ##

Each row of the training data contains a click record, with the following features.

- `ip`: ip address of click.
- `app`: app id for marketing.
- `device`: device **type** id of user mobile phone (e.g., iphone 6 plus, iphone 7, huawei mate 7, etc.)
- `os`: os version id of user mobile phone
- `channel`: channel id of mobile ad publisher
- `click_time`: timestamp of click (UTC)
- `attributed_time`: if user download the app for after clicking an ad, this is the time of the app download
- `is_attributed`: the target that is to be predicted, indicating the app was downloaded

Note that `ip`, `app`, `device`, `os`, and `channel` are encoded.

The test data is similar, with the following differences:

- `click_id`: reference for making predictions
- `is_attributed`: not included

## Checking baseline conversion (download ratio) ##

````python
conversion = df[['is_attributed', 'click_time']].pivot_table(columns='is_attributed', aggfunc='count')
conversion['conversion (%)'] = (100*conversion[1]/(conversion[1]+conversion[0])).round(3)
conversion
````

## Distribution of encoded variables ##

Now we are going to check distributions for the encoded columns. What we want to understand is if there is a skewed behaviour in their occurence numbers. 

First we need the understand the quantity of unique values for each variable. Also, we are going to tranform them into categorical variables.

````python
# Checking quantity of ids of ip, app, device, os and channel, which are encoded #
encoded_cols = ['ip', 'app', 'device', 'os', 'channel']
for col in encoded_cols:
    df[col] = df[col].astype('category')
    
quantities = [len(df[col].unique()) for col in encoded_cols]

print('Quantity of ids for encoded columns:')
for i in range(len(encoded_cols)):
    print('{}:     \t {}'.format(encoded_cols[i], quantities[i]))
````

### ip: ###

`ip` frequency is very skewed, with a very small quantity of `ip` repeating for more than 10 times in a very small interval of time. This `ip` may not be unique for each device, but we would also expect a similar behaviour from a fraudulent activity.

````python
ip_agg = df.groupby(['ip']).agg({ 
                                'attributed_time': ['count'],
                                'ip': ['count'] })
ip_agg = ip_agg[['ip','attributed_time']]

ip_agg['conversion (%)'] = (100*ip_agg['attributed_time']
                            /ip_agg['ip']).round(3)

ip_agg = ip_agg.sort_values(by=[('ip', 'count')], ascending=True)
ip_agg_sort = np.array(ip_agg[('ip','count')])
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set(title='Count of clicks')
line = ax.plot(ip_agg_sort)
ax.set_yscale('log')
````

The rolling average shows that conversion rates fall for common `ip`, which at least indicates that most of those `ip` may be fraudulent.

````python
ip_rolling_conversion = np.array(pd.rolling_mean(ip_agg['conversion (%)'], window=5000))
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set(title='Rolling conversion (%)')
line = ax.plot(ip_rolling_conversion)
````

Top two `ip` have very nice conversion, compared to baseline, which strengthens the theory that `ip` may not be unique, in some instances, or this can be a fraud strategy. 

Some other top `ip` not so much. Although we can see some clear patterns in those `ip`, we have many different `ip` in the dataset, and probably `test.csv` is going to have many other new `ip`. This only proves that, although `ip` may be used to increase highly prediction score, it needs to be used with caution.

````python
ip_agg_extremes = pd.concat([ip_agg.head(10), ip_agg.tail(10)])
print(ip_agg_extremes)
ip_agg_extremes[('ip', 'count')].plot.bar(figsize=(12,3))
plt.show()
````

We are going to visualize hourly click and conversion rates to check for patterns. Fraudulent activity may be associated with a large quantity of clicks and declining conversion rates.

At first we can see a clear pattern repeating itself:

````python
df['click_time_hour']=df['click_time'].dt.round('H')  

df[['click_time_hour','is_attributed']].groupby(['click_time_hour']).count().plot()
plt.title('Clicks per hour')
plt.show()
````

By grouping by hour we can understand better the pattern. There's a highly unexpected clicking activity late during the night, with observed peak on 4h, and there's a decrease on clicking activity at eve, with observed valley on 20h.

````python
df['click_hour'] = df['click_time'].dt.hour
df[['click_hour','is_attributed']].groupby(['click_hour']).count().plot()
plt.title('Clicks per hour (group by hour)')
plt.show()
````

Checking activity of `ip` with high click frequency and 0 conversion we can see a similar behaviour with the exception that there's a peak of activity at noon.

````python
suspicious_ip = ip_agg[(ip_agg[('ip','count')]>=10) & (ip_agg['conversion (%)']==0)].index
df[df['ip'].isin(suspicious_ip)][['click_hour','is_attributed']].groupby(['click_hour']).count().plot()
plt.title('Clicks per hour by suspicious ip (group by hour)')
plt.show()
````

## Download delay ##

We raised the hyphothesis that `ip` may be a shared element between different devices, but we also stated that there's a chance that eventual conversions on high-frequency `ip` may be a fraudulent strategy, maybe to fool TalkingData's ip blacklist model.

If a robot is generating clicks with the same `ip`, and eventually downloading the app, we may notice some pattern in the delay between `click_time` and `attributed_time`.

From the analysis of download delay we can see that there are some very quick downloads, which may be a fraudulent strategy, if this information was tracked correctly.

````python
df['download_delay']= (df['attributed_time']-df['click_time']).dt.total_seconds()

ip_agg_delay = df.groupby(['ip']).agg({ 
                                'attributed_time': ['count'],
                                'ip': ['count'],
                                'download_delay': ['mean']})

ip_agg_delay = ip_agg_delay[ip_agg_delay[('attributed_time', 'count')] > 0]
ip_agg_delay = ip_agg_delay.sort_values(by=[('ip', 'count')], ascending=True)

ip_agg_delay[('download_delay', 'mean')].describe()
````
